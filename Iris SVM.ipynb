```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_Iris = pd.read_csv('Iris.csv')
```

```python
df_Iris
```

```python
Species = df_Iris.Species.unique()
```

```python
Species =Species.tolist()
```

```python
count = [0]*len(Species)
```

```python
for i in Species:
    for j in df_Iris['Species']:
        if ( i == j ):
            count[Species.index(i)] += 1
```

```python
count
```

```python
print(Species)
```

```python
print(count)
```

```python

plt.figure(figsize=(20,20))
plt.title("Pie Chart of Species")
plt.pie(count,labels=Species,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df_Iris,hue='Species')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_Iris.iloc[:,1:5].corr(),annot=True,fmt='.0%')
```

```python
from sklearn.model_selection import train_test_split
X = df_Iris.iloc[:,1:5]
Y = df_Iris.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, Y , test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, matthews_corrcoef
print(confusion_matrix(y_test,preds))
```

```python
print(classification_report(y_test,preds))
```
